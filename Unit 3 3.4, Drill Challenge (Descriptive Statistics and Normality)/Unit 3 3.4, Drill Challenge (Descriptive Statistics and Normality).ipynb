```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

binomial = np.random.binomial(30, 0.5, 100)
plt.hist(binomial, bins=15, color='c')
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
print("Very dependent on the random variable, but distribution is generally symmetric.")
```

```python
beta = np.random.beta(3, 5, 100)
plt.hist(beta, bins=20, color='m')
plt.axvline(beta.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(beta.mean() + beta.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(beta.mean() - beta.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
print("Very dependent on the random variable, but distribution is lightly skewed to the right.")
```

```python
pareto = np.random.pareto(1000, 100)
plt.hist(pareto, bins=15, color='c')
plt.axvline(pareto.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pareto.mean() + pareto.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(pareto.mean() - pareto.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
print("Very dependent on the random variable, but distribution is heavily skewed to the right.")
```

```python
rayleigh = np.random.rayleigh(1000, 100)
plt.hist(rayleigh, bins=15, color='c')
plt.axvline(rayleigh.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rayleigh.mean() + rayleigh.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(rayleigh.mean() - rayleigh.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
print("Very dependent on the random variable, but distribution is heavily skewed to the right. Occasional symmetric.")
```

```python
logseries = np.random.logseries(0.6, 100)
plt.hist(logseries, bins=8, color='c')
plt.axvline(logseries.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logseries.mean() + logseries.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(logseries.mean() - logseries.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
print("Very dependent on the random variable, but distribution is heavily skewed to the right.")
```

```python
power = np.random.power(500, 100)
plt.hist(power, bins=8, color='c')
plt.axvline(power.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(power.mean() + power.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(power.mean() - power.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
print("Distribution is heavily skewed to the left.")
```

```python
# Additional Distribution
rand1 = np.random.normal(5, 0.5, 100)
rand2 = np.random.normal(10, 1, 100)
rand3 = rand1 + rand2
plt.hist(rand3, bins=10, color='c')
plt.axvline(rand3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean() - rand3.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
print("Distribution is mostly symmetrical, with slight skew.")
```
