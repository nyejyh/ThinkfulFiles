```python
# Unit 2 Challenge by Jason Huang
# Using 2016 American Community Survey for database

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Read the JSON.
data = pd.read_json('2016_detailed_tables.json')

# This is an index of coded names.
names = data.index
```

```python
# Create an empty data frame with named rows.
df = pd.DataFrame(index=names)

# Create empty lists to fill for specific information.
gender_list = []
age_list = []

# Cycle and add our columns one at a time for each row.
# Each row has the following strings: label, group, concept.
# All other data are not important.
for i in names:
    
    # Split the label list
    id_list = re.split("!!",data["variables"][i]["label"])
    #print(id_list)
    
    # Build the gender list
    if "Male" in id_list:
        gender_list.append("M")
    elif "Female" in id_list:
        gender_list.append("F")
    else:
        gender_list.append(np.nan)

    # Build the age range per row
    if len(id_list) > 3 and ("Male" in id_list or "Female" in id_list):
        #print(id_list[-1])
        age_list.append(id_list[-1])
    else:
        age_list.append(np.nan)
    
df["Gender"] = gender_list
df["Age"] = age_list
```

```python
# Make categorical data based on humans only
# Not able to get ride of anything unrelated to age
humans_only = df.dropna()
categorical = pd.get_dummies(humans_only, columns=["Age"])
categorical
```

```python
# Compare the number of males versus females, where Males count on the left, Females on the right.
# Seems like they are about the same.
gender_count = []
for i in gender_list:
    if i == "M":
        gender_count.append(1)
    elif i == "F":
        gender_count.append(2)
    else:
        gender_count.append(0)
plt.hist(gender_count, bins=5, color='green')
plt.title("Gender Counts")
plt.ylabel("Number of Counts")
plt.xlabel("N/A = 0, M = 1, F = 2")
plt.show()
```

```python
# Make a plot of genders from list, not count.
# It seems like we can tell which of the data has gender implications.
plt.plot(gender_list)
plt.show()
```

```python
# Make a bargraph of number of available categories based on data from "label".
plt.hist(categorical, bins=1)
```
