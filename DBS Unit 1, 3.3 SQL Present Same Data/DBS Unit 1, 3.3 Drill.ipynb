```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load dataset from Riot Games
df = pd.read_csv('matchinfo.csv')
print(df.head(n=5))

# Choose one variable and plot it four different ways
# Create variables 
TSMgamelength = df.loc[(df['blueTeamTag']=='TSM'),'gamelength']

# Create histogram
plt.hist(TSMgamelength, color='green')
plt.xlabel('Length of Games in Minutes')
plt.ylabel('Number of Games')
plt.title('Plot 1: Length of Games for Team TSM')
plt.show()

# Create boxplot
ax = sns.boxplot(y='gamelength',data=df)
plt.title('Plot 2: Range of Length of Games for Team TSM')
plt.show()

# KDE on univariate distribution
sns.distplot(TSMgamelength)
plt.title('Plot 3: Length of Games for Team TSM')
plt.show()

# KDE with bandwidth parameters
sns.kdeplot(TSMgamelength)
sns.kdeplot(TSMgamelength, bw=.9, label="bw: 0.9")
sns.kdeplot(TSMgamelength, bw=5, label="bw: 5")
plt.title('Plot 4: Length of Games for Team TSM with Bandwidth')
plt.legend()
plt.show()
```

```python
# Choose two continuous variables and plot it three different ways
# Load dataset from Riot Games
df2 = pd.read_csv('gold.csv')
print(df2.head(n=5))

# Choose two continuous variables, and plot them in three different ways.

# Create scatterplot
sns.jointplot(x='min_10', y='min_20', data=df2)
plt.title('Plot 1: Gold for Matches at Minute 10 vs Minute 20')
plt.show()

# Create Hexbin plot
with sns.axes_style("white"):
    sns.jointplot(x=df2['min_10'], y=df2['min_80'], kind="hex", color="k")
plt.title('Plot 2: Gold for Matches at Minute 10 vs Minute 80')
plt.show()
    
# Kernel density estimation
g = sns.jointplot(x="min_10", y="min_80", data=df2, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
plt.title('Plot 3: Gold for Matches at Minute 10 vs Minute 80')
plt.show()
```

```python
# Choose one continuous variable and one categorical variable and plot it six different ways

# Create strip plot
sns.stripplot(x=df['blueTeamTag'], y=df['gamelength'], data=df)
plt.title('Plot 1: Game Length per Team on Blue Side')
plt.show()

# Create strip plot
sns.boxplot(x=df['blueTeamTag'], y=df['gamelength'], data=df)
plt.title('Plot 2: Game Length per Team on Blue Side')
plt.show()

# Create violin plot
sns.violinplot(x=df['blueTeamTag'], y=df['gamelength'], data=df)
plt.title('Plot 3: Game Length per Team on Blue Side')
plt.show()

# Create bar plot
sns.barplot(x=df['blueTeamTag'], y=df['gamelength'], data=df)
plt.title('Plot 4: Game Length per Team on Blue Side')
plt.show()

# Create Point Plot
sns.pointplot(x=df['blueTeamTag'], y=df['gamelength'], data=df)
plt.title('Plot 5: Game Length per Team on Blue Side')
plt.show()

# Create count plot data
sns.countplot(y=df['gamelength'], data=df, palette="Greens_d")
plt.title('Plot 6: Game Length per Team on Blue Side')
plt.show()
```
