```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline

def population(pop1, pop2):
    plt.hist(pop1, alpha=0.5, label='Population 1')
    plt.hist(pop2, alpha=0.5, label='Population 2')
    plt.legend(loc='upper right')
    plt.show()
    
def sampling(sample1, sample2):
    plt.hist(sample1, alpha=0.5, label='sample 1')
    plt.hist(sample2, alpha=0.5, label='sample 2')
    plt.legend(loc='upper right')
    plt.show()
    
def t_test(sample1, sample2):
    # Calculate the T-test
    # Compute the difference between the two sample means.
    diff = sample2.mean() - sample1.mean()
    print('The difference between two sample mean is {}.'.format(diff))
    # Create the arrays to calculate the standard error of difference.
    size = np.array([len(sample1), len(sample2)])
    sd = np.array([sample1.std(), sample2.std()])
    print('The arrary for size is {}.'.format(size))
    print('The array for standard deviation is {}.'.format(sd))
    # Standard Error equation is squared standard deviations divided by sample size, summed, then square rooted.
    diff_se = (sum((sd ** 2) / size)) ** 0.5
    print('The standard error of this dataset is {}.'.format(diff_se))
    # T-Value is difference between means divided by Standard Error
    print('The T-value of this dataset is {}.'.format(diff / diff_se))
```

```python
# Starting values for populations
# pop1 = np.random.binomial(n, p-value, size) 
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

population(pop1, pop2)
```

```python
# Made with binomially distributed variable and a sample size of 100
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

sampling(sample1, sample2)

# Calculate the T-test
t_test(sample1, sample2)
```

```python
# Increase the size of samples from 100 to 1000.
# Because size of samples increased, the y-axis should increase by 10 times.
sample3 = np.random.choice(pop1, 1000, replace=True)
sample4 = np.random.choice(pop2, 1000, replace=True)

sampling(sample3, sample4)

# Calculate the T-test
t_test(sample3, sample4)
```

```python
# Decrease the size of samples from 100 to 20.
# Because size of samples  the y-axis should decrease to less than 10.
sample5 = np.random.choice(pop1, 20, replace=True)
sample6 = np.random.choice(pop2, 20, replace=True)

sampling(sample5, sample6)

# Calculate the T-test
t_test(sample5, sample6)
```

```python
# Change values for pop1 to 0.3
# This makes the pop1 increase similarity to pop2 since probability value increased.
# pop1 = np.random.binomial(n, p-value, size) 
pop3 = np.random.binomial(10, 0.3, 10000)
pop4 = np.random.binomial(10, 0.5, 10000)

population(pop3, pop4)
```

```python
# Change values for pop1 to 0.4
# This makes the pop1 increase similarity to pop2 even more since probability value increased.
# Increase of probability value means failing to reject null hypothesis.
# pop1 = np.random.binomial(n, p-value, size) 
pop5 = np.random.binomial(10, 0.4, 10000)
pop6 = np.random.binomial(10, 0.5, 10000)

population(pop5, pop6)
```

```python
# Change the distribution of populations to power.
# power(a[, size])
pop7 = np.random.power(10, 10000)
pop8 = np.random.power(50, 10000)

population(pop7, pop8)
```

```python
# Decrease the size of samples from 100 to 20.
# Because size of samples  the y-axis should decrease to less than 10.
sample7 = np.random.choice(pop7, 100, replace=True)
sample8 = np.random.choice(pop8, 100, replace=True)

sampling(sample7, sample8)

# Calculate the T-test
t_test(sample7, sample8)

print("The distribution of power gives a similar graph like population.")
```
