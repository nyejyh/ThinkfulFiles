```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import statistics
%matplotlib inline

df = pd.read_csv('plane_crashes_data.csv')
print(df)
```

```python
# Calculate the mean of the fatalities for the years 1980 to 2009 inclusive.
# Dataframe goes from 1950 to 2009.
eighties = df.loc[lambda df: df['year'] > 1979, :]
eighties_mean = eighties.mean()
print(eighties_mean)
```

```python
# Calculate which month that has the most plane crashes out of 1950 to 2009.
# Mode is the most number of times something appears.
# Month of 1 means it is January.
mode = statistics.mode(df['month'])
print('The mode of this dataset is {}.'.format(mode))
```

```python
# Calculate the highest number of fatalities in a single crash.
# .idxmax() returns index labels.
fatal_high = df.loc[df['fatalities'].idxmax()]
print(fatal_high)
```

```python
# Calculate the percentage of crashed flights that is 1 or more fatalities.
fatally_high = 0
for i in df['fatalities']:
    if i >= 1:
        #print(df['fatalities'][i])
        fatally_high += 1
fatal_percent = fatally_high / len(df['fatalities'])
print(fatal_percent * 100)
```

```python
# Calculate the average percent of passengers & crew killed per plane crash for entire period.
percent_sum = 0
for i in df['fatalities']:
    aboard = df['aboard'][i]
    fatal = df['fatalities'][i]
    percent = fatal / aboard
    percent_sum += percent
    #print(percent)
    
print(len(df['fatalities']))
print(percent_sum)
print(percent_sum / len(df['fatalities']))
```

```python
# Testing if different by rounding
from decimal import Decimal
percent_sum = 0
for i in df['fatalities']:
    aboard = df['aboard'][i]
    fatal = df['fatalities'][i]
    percent = Decimal(fatal / aboard)
    percent_sum += round(percent,2)
    #print(percent)
    
print(len(df['fatalities']))
print(percent_sum)
print(percent_sum / len(df['fatalities']))
```
