```python
# Unit 2 Challenge by Jason Huang
# Use dataset given as un.csv

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Read the CSV file.
df = pd.read_csv("un.csv")
df
```

```python
print(df.columns)
```

```python
# Using layered plot.
# Compare the life expectancy of males to females worldwide.
# Green line stands for females, in which they have higher life expectancy generally.
plt.plot(df["lifeMale"], color = "Red")
plt.plot(df["lifeFemale"], color = "Green")
plt.ylabel("Years in Life")
plt.xlabel("Country by Alphabetical Order")
plt.title("Life Expectancy of Males VS Females")
plt.show()
```

```python
# Using scatter plots.
# Compare the GDP per capita of contries to infant mortality rate.
# The higher the GDP Per Capita goes, the lower the Infant Mortality Rate gathers.
# The lower the GDP Per Capita goes (below 5000), the higher the infant mortality rate becomes.
plt.scatter(x=df["infantMortality"], y=df["GDPperCapita"])
plt.ylabel("GDP Per Capita")
plt.xlabel("Infant Mortality Rate")
plt.title("Relationship between GDP Per Capita and Infant Mortality Rate")
plt.show()
```

```python
# Using subplots for scatterplots.
# For both males and females, higher education means higher chance of contraception safety.
# At lower education for both sexes, contraception rate is lower.
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x = df["educationMale"], y = df["contraception"], color="red")
plt.ylabel("Contraception Rate")
plt.xlabel("Percentage of Males Educated")
plt.title("Contraception Rate and Education for Male")

plt.subplot(1, 2, 2)
plt.scatter(x = df["educationFemale"], y = df["contraception"], color="green")
plt.ylabel("Contraception Rate")
plt.xlabel("Percentage of Females Educated")
plt.title("Contraception Rate and Education for Females")

plt.tight_layout()
plt.show()
```

```python
# Using historgrams.
# Any missing values have been converted to value of "0", so ignore the counts of zeros.
# This is valid logic because zero illiteracy rate in any country should not be possible.
# There are fewer countries with high illiteracy rate as you proceed positively along x axis.
# There are more countries with lower illiteracy rate in general.
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
x = df["illiteracyMale"].fillna(0)
plt.hist(x, bins=10, color = "red")
plt.ylabel("Counts")
plt.xlabel("Illiteracy Rate")
plt.title("Number of Worldwide Illiteracy Rates for Males")

plt.subplot(1, 2, 2)
y = df["illiteracyFemale"].fillna(0)
plt.hist(x, bins=10, color = "Green")
plt.ylabel("Counts")
plt.xlabel("Illiteracy Rate")
plt.title("Number of Worldwide Illiteracy Rates for Females")

plt.tight_layout()
plt.show()
```
